```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("List of Countries by Sugarcane Production.csv")
df.head()
```

```python
df.drop('Unnamed: 0' , axis = 1 , inplace = True)
```

```python
df.head()
```

```python
df.rename(columns = {"Production (Tons)" : "Production(Tons)"},inplace = True)
df.rename(columns = {"Production per Person (Kg)" : "Production_per_Person(Kg)"},inplace = True)
df.rename(columns = {"Acreage (Hectare)" : "Average(Hectare)"},inplace = True)
df.rename(columns = {"Yield (Kg / Hectare":"Yield(Kg/Hectre)"},inplace = True)
```

```python
df.head()
```

```python
df["Production(Tons)"] = df["Production(Tons)"].astype(str).str.replace(",", "")
df["Average(Hectare)"] = df["Average(Hectare)"].astype(str).str.replace(",", "")
df["Production_per_Person(Kg)"] = df["Production_per_Person(Kg)"].astype(str).str.replace(".", "")
df["Production_per_Person(Kg)"] = df["Production_per_Person(Kg)"].astype(str).str.replace(",", "")
df["Yield (Kg / Hectare)"] = df["Yield (Kg / Hectare)"].astype(str).str.replace(".", "")
df["Yield (Kg / Hectare)"] = df["Yield (Kg / Hectare)"].astype(str).str.replace(",", "")
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["Production(Tons)"] = df["Production(Tons)"].astype(float)
df["Average(Hectare)"] = df["Average(Hectare)"].astype(float)
df["Yield (Kg / Hectare)"] = df["Yield (Kg / Hectare)"].astype(float)
```

```python
df.head()
```

```python
df['Continent'].value_counts()
```

```python
plt.figure(figsize = (10,10))
plt.subplot(2,2,1)
sns.distplot(df["Production(Tons)"])
plt.subplot(2,2,2)
sns.distplot(df["Production_per_Person(Kg)"])
plt.subplot(2,2,3)
sns.distplot(df["Average(Hectare)"])
plt.subplot(2,2,4)
sns.distplot(df["Yield (Kg / Hectare)"])
```

```python
#checking outliers
plt.figure(figsize = (20,50))
plt.subplot(2,2,1)
sns.boxplot(df["Production(Tons)"])
plt.subplot(2,2,2)
sns.boxplot(df["Production_per_Person(Kg)"])
plt.subplot(2,2,3)
sns.boxplot(df["Average(Hectare)"])
plt.subplot(2,2,4)
sns.boxplot(df["Yield (Kg / Hectare)"])
```

```python
plt.figure(figsize = (20,20))
plt.pie(df['Production(Tons)'], labels=df['Continent'], autopct='%.0f%%') 
plt.show() 
```

```python
#continent with highest production
plt.figure(figsize = (10,10))
sns.barplot(y = df['Production(Tons)'] , x = df['Continent']) 
plt.show() 
```

```python
#continent with highest land
plt.figure(figsize = (10,10))
sns.barplot(y = df['Production(Tons)'] , x = df['Continent']) 
plt.show() 
```
